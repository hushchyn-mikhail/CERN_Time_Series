```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Time Series Generation

#Constant

```python
K=104
N = 3000

series = []

X = np.array(range(1, 105))
for i in range(0, N):
    mean = np.random.random_integers(20, 100, 1)[0]
    sample = np.random.normal(mean, 5, X.shape[0])
    
    cut = np.random.random_integers(10, X.shape[0], 1)[0]
    proba = np.random.random(1)[0]
    holes = np.random.binomial(1, proba, X.shape[0])
    mask = 1.*(X <= cut)*holes
    
    ser = sample*mask
    series.append(ser)

series = np.array(series)
new_data = pd.DataFrame(series)
new_data.columns = [str(i) for i in range(1, X.shape[0]+1)]
```

```python
new_data.to_csv('simulation.csv')
```

```python
X = np.array(range(1, 105))
N=20
plt.figure(figsize=(15, 1*N))
for i in range(0, N):
    serie = new_data.irow([i]).values[0]
    plt.subplot(int(N/2)+1, 2, i+1)
    plt.plot(X, serie)
plt.show()
```

#Linear

```python
K=104
N = 3000

series = []

X = np.array(range(1, 105))
for i in range(0, N):
    k = np.random.normal(-3, 3, 1)[0]
    b = np.random.normal(150, 20, 1)[0]
    sample = k*X + b + np.random.normal(0, 45, X.shape[0])
    
    proba = np.random.random(1)[0]
    holes = np.random.binomial(1, proba, X.shape[0])
    mask = 1.*(sample > 0)*holes
    
    ser = sample*mask
    series.append(ser)

series = np.array(series)
new_data = pd.DataFrame(series)
new_data.columns = [str(i) for i in range(1, X.shape[0]+1)]
```

```python
new_data.to_csv('simulation.csv')
```

```python
X = np.array(range(1, 105))
N=20
plt.figure(figsize=(15, 1*N))
for i in range(0, N):
    serie = new_data.irow([i]).values[0]
    plt.subplot(int(N/2)+1, 2, i+1)
    plt.plot(X, serie)
plt.show()
```

#Gaus

```python
K=104
N = 3000

series = []

X = np.array(range(1, 105))
for i in range(0, N):
    mean = np.random.random_integers(20, 100, 1)[0]
    sigma = np.random.normal(10, 10, 1)[0]
    intensity = np.random.normal(70, 20, 1)[0]
    sample = intensity*np.exp(-(1.*(X - mean)**2)/sigma**2) + np.random.normal(-5, 2, X.shape[0])
            
    sample = sample
    
    proba = np.random.random(1)[0]
    holes = np.random.binomial(1, proba, X.shape[0])
    mask = 1.*(sample > 0)*holes
    
    ser = sample*mask
    series.append(ser)

series = np.array(series)
new_data = pd.DataFrame(series)
new_data.columns = [str(i) for i in range(1, X.shape[0]+1)]
```

#Gaus + pikes

```python
K=104
N = 3000

series = []

X = np.array(range(1, 105))
for i in range(0, N):
    mean = np.random.random_integers(20, 100, 1)[0]
    sigma = np.random.normal(10, 10, 1)[0]
    intensity = np.random.normal(70, 20, 1)[0]
    sample = intensity*np.exp(-(1.*(X - mean)**2)/sigma**2) + np.random.normal(-5, 2, X.shape[0])
    
    pikes = 200*np.random.binomial(1, 0.01, X.shape[0]) + \
            500*np.random.binomial(1, 0.01, X.shape[0]) + \
            1000*np.random.binomial(1, 0.01, X.shape[0]) + \
            10000*np.random.binomial(1, 0.01, X.shape[0])
            
    sample = sample + pikes
    
    proba = np.random.random(1)[0]
    holes = np.random.binomial(1, proba, X.shape[0])
    mask = 1.*(sample > 0)*holes
    
    ser = sample*mask
    series.append(ser)

series = np.array(series)
#series = np.log(series+1)
new_data = pd.DataFrame(series)
new_data.columns = [str(i) for i in range(1, X.shape[0]+1)]
```

```python
new_data.to_csv('simulation.csv')
```

```python
X = np.array(range(1, 105))
N=20
plt.figure(figsize=(15, 1*N))
for i in range(0, N):
    serie = new_data.irow([i]).values[0]
    plt.subplot(int(N/2)+1, 2, i+1)
    plt.plot(X, serie)
plt.show()
```
